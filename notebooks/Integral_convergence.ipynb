```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

# A very simple notebook to check order of convergence for numerical integral

```python
# Define function to be integrated, integration limits and reference solution.
f = lambda x: x + x**2 + x**3
lower_limit = 0
upper_limit = 1
reference = 13/12


f = lambda x: np.abs(np.sin(x*np.pi))
lower_limit = 0
upper_limit = 2
reference = 4/np.pi


# Calculate numerical integral with a range of different bin sizes
Nbins = np.array([2, 4, 8, 16, 32, 64, 128, 256, 512, 1024])
result = np.zeros(len(Nbins))

for i, N in enumerate(Nbins):
    # Cell faces, cell size, and cell centers
    xf = np.linspace(lower_limit, upper_limit, N+1)
    dx = xf[1] - xf[0]
    xc = xf[:-1] + dx/2
    # Calculate integral with midpoint rule
    result[i] = dx*np.sum(f(xc))
```

```python
# plot convergence
plt.plot(Nbins, np.abs(result - reference))
plt.plot(Nbins, 1/Nbins**2, '--', c = 'k', label = '$\sim 1/N^2$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('N_{bin}')
plt.ylabel('Error')
plt.legend()
```
