```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
f = lambda x: x + x**2 + x**3
reference = 13/12
lower_limit = 0
upper_limit = 1



Nbins = np.array([2, 4, 8, 16, 32, 64, 128, 256, 512, 1024])
result = np.zeros(len(Nbins))

for i, N in enumerate(Nbins):
    # Cell faces, cell size, and cell centers
    xf = np.linspace(lower_limit, upper_limit, N+1)
    dx = xf[1] - xf[0]
    xc = xf[:-1] + dx/2
    # Calculate integral with midpoint rule
    result[i] = dx*np.sum(f(xc))
```

```python
plt.plot(Nbins, np.abs(result - reference))
plt.plot(Nbins, 1/Nbins**2, '--', c = 'k', label = '$\sim 1/N^2$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('N_{bin}')
plt.ylabel('Error')
plt.legend()
```
