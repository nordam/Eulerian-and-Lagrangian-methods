```python
Z = np.load('../results/Case4_K_A_lagrangian_Nparticles=50000_dt=600.0_Z_0000.npy')
D = np.load('../results/Case4_K_A_lagrangian_Nparticles=50000_dt=600.0_D_0000.npy')

NtL, Np = Z.shape
print(NtL, Np)
```

```python
Nk = 64 

Dmin = 1e-8
Dmax = 2e-4

Rf = np.logspace(np.log10(Dmin), np.log10(Dmax), Nk+1)
Rc = np.sqrt(Rf[:-1]*Rf[1:])

fig = plt.figure(figsize = (9, 5))

for it in range(NtL):
#for it in [-1,]:

    h, _ = np.histogram(D[it,:], bins = Rf)
    h = h/np.sum(h)

    plt.plot(Rc, h)


plt.legend()
plt.xscale('log')
plt.xlabel('Particle size [m]')
```

```python
Zmax = 50
zf, dz = np.linspace(0, Zmax, 101, retstep=True)
zc = zf[:-1] + dz/2

for it in [0, 5, 10, -1]:
    C, _ = np.histogram(Z[it,:], bins=zf)
    plt.plot(C, zc)
    
fig = plt.figure(figsize = (9, 5))
```

```python
time = np.linspace(0, 48*3600, NtL)
remaining = np.sum((Z>0) & (D > 0) & (D < 1e-6), axis=1)

fig = plt.figure(figsize = (4,3))
plt.plot(time/3600, remaining / remaining[0])
plt.xlabel('Time')
plt.ylabel('Fraction')
plt.ylim(-0.0, 1.03)
plt.tight_layout()
plt.savefig('remaining_zero.png', dpi=240)
```

```python
Np = 20000
Nk = 32

Dmin = 1e-8
Dmax = 1e-4

Rf = np.logspace(np.log10(Dmin), np.log10(Dmax), Nk+1)
Rc = np.sqrt(Rf[:-1]*Rf[1:])

D = lognorm(0.5, scale=3e-6).rvs(size=Np)

h, _ = np.histogram(D, bins = Rf)
h = h/np.sum(h)

plt.bar(Rc, h, width=0.9*(Rf[1:]-Rf[:-1]))
plt.xlim(Dmin, Dmax)
plt.xscale('log')
```
